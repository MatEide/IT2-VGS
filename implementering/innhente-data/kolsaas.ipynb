```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur[1][0] #-> gir meg den f√∏rste tiden 0
```

```python
tur
```

```python
## oppgave: plott time (tid) p√• x-aksen mot altitude (h√∏yde) p√• y-aksen

x = []
y = []

for maalepunkt in tur[1:]:
    tid = float(maalepunkt[1])
    x.append(tid)
    h√∏yde = float(maalepunkt[3])
    y.append(h√∏yde)

plt.scatter(x,y)
plt.show()



```
